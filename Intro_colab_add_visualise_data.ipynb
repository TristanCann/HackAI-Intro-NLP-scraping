##Let's do a class poll: What type of text analysis is most interesting / exciting to you?

```python
import pandas as pd # let's import the pandas library, for data analysis

# we will modify these numbers based on the class
data = {'Product reviews sentiment': [3], 'Analyse parliamentary debates': [2], 'Mine Corporate Social Responsibility in reports': [3],
        'Literature trends': [1], 'Public conversation in Social media': [5], 'Students academic writing': [1], 'Rhetoric in crowdfunding narratives': [1] } # create a dictionary of data about what we are most interested in
```

```python
data # check the data
```

```python
df = pd.DataFrame(data, ["Course attendees"])
df
```

```python
# get a specific value from the dataframe
df.loc["Course attendees", "Product reviews sentiment"]
```

```python
# display number only - this is necessary due to the way that the latest version of numpy displays numbers, we can ask it to print like an older version to just get the number
import numpy as np
np.set_printoptions(legacy='1.25')

df.loc["Course attendees", "Product reviews sentiment"]
```

```python
type(df.columns)
```

```python
df.columns.values
```

```python
df.loc["Course attendees"].values
```

```python
df.iloc[0]
```

```python
import matplotlib.pyplot as plt #visualisation library

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df.columns.values
sizes = df.loc["Course attendees"].values

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
       startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

# Let's do another poll: what type of text data interests you?

*Also consider if we can look at it together in the class*

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

class_responses = ['pre_existing_data', 'webpage', 'pdf', 'share with class']
counts = [1, 5, 4, 1]
#bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:orange', 'tab:green']

ax.bar(class_responses, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('number of students')
ax.set_title('Class text data interests')


plt.show()
```
